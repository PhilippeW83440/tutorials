## SVD decomposition 

```python
# M = U s V' with U and V orthogonal and s diagonal

M = np.array([[1,2,3],[4,5,6]])
print(M)
U, s, W = np.linalg.svd(M) # M = U s V' = U s W
print(U)
print(s)
print(np.dot(W, W.T)) # W is orthogonal: Identity = W * W'
```

```python
S = np.zeros_like(M, dtype='float')
S[:U.shape[0], :U.shape[1]] = np.diag(s)
M_recons = np.dot(U,np.dot(S,W)) # M_recons = U * S * W
print(M_recons)
```

## Copy

```python
A=np.array([[0,2], [3,4]])
B=A
B[0,0]=33
print(A)
```

```python
A=np.array([[0,2], [3,4]])
B=A.copy()
B[0,0]=33
print(A)
```

```python
n,m=A.shape
B=A.reshape((n*m,))
print(B)
B=A.reshape((n*m,1))
print(B)
```

## Concatenate (vstack, hstack) and Repeat (tile)

```python
A=np.array([[1,2], [3,4]])
print(A)
print(np.tile(A, (1,2)))
```

```python
B=np.array([[5,6]])
print(B)
print(np.concatenate((A,B), axis=0))
print(np.vstack((A,B)))
```

```python
print(np.concatenate((A,B.T), axis=1))
print(np.hstack((A,B.T)))
```
